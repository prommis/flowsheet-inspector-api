## Imports
Get some demo flowsheets from the tests.

```python
from idaes_fi.structfs.tests import demo_flowsheet, demo_flowsheet_structured
```

Get the decorator used to wrap a single main function.

```python
from idaes_fi.structfs import fi_main
```

```python
@fi_main(name="usage-demo2")
def main():
    m = demo_flowsheet.build_flowsheet()
    demo_flowsheet.set_scaling(m)
    demo_flowsheet.set_dof(m)
    demo_flowsheet.initialize_flowsheet(m)
    solver = demo_flowsheet.get_solver()
    result = demo_flowsheet.solve_flowsheet(m, solver, stee=True)
    demo_flowsheet.display_results(m)

    return m, result
```

```python
main()
```
